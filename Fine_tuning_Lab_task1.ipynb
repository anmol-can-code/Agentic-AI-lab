```python
from datasets import load_dataset

dataset = load_dataset("openwebtext")
dataset
```

```python
small_dataset = dataset["train"].shuffle(seed=42).select(range(10000))
```

```python
print(small_dataset[0]["text"][:500])
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "distilgpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
def tokenize_function(examples):
    outputs = tokenizer(
        examples["text"],
        truncation=True,
        padding="max_length",
        max_length=128
    )

    # 🔥 CRITICAL FIX
    outputs["labels"] = outputs["input_ids"].copy()

    return outputs
```

```python
tokenized_dataset = small_dataset.map(
    tokenize_function,
    batched=True,
    remove_columns=["text"]
)
```

```python
tokenized_dataset = tokenized_dataset.train_test_split(test_size=0.1)

train_dataset = tokenized_dataset["train"]
eval_dataset = tokenized_dataset["test"]
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    do_eval=True,
    learning_rate=5e-5,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=2,
    weight_decay=0.01,
    save_total_limit=2,
    logging_dir="./logs",
    logging_steps=100,
    report_to="none",
    fp16=True
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
)
```

```python
trainer.train()
```

```python
eval_results = trainer.evaluate()
eval_results
```

```python
import math
math.exp(eval_results["eval_loss"])
```

```python
prompt = "Artificial intelligence is rapidly changing"

inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

outputs = model.generate(
    **inputs,
    max_length=60,
    do_sample=True,
    top_k=50,
    top_p=0.95
)

tokenizer.decode(outputs[0], skip_special_tokens=True)
```
